```python
deck_a= ti_df.loc[ti_df['deck'] == 'A']
deck_b= ti_df.loc[ti_df['deck'] == 'B']
deck_c= ti_df.loc[ti_df['deck'] == 'C']
deck_d= ti_df.loc[ti_df['deck'] == 'D']
deck_e= ti_df.loc[ti_df['deck'] == 'E']
deck_f= ti_df.loc[ti_df['deck'] == 'F']
deck_g= ti_df.loc[ti_df['deck'] == 'G']
h = sns.FacetGrid(deck_a, hue = 'survived')
h.map(plt.hist, 'who')
```
